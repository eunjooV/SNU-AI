```python
import numpy as np
from numpy import sin, cos
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

## 1. sin‚Å°(ùë•+ùúã)  ÏôÄ cos‚Å°(ùë•) Î•º Ìïú plotÏúÑÏóê Í∑∏Î¶¨ÏãúÏò§

Ï°∞Í±¥
* Figure Ï†úÎ™©: Signal Processing
* XÍ∞í Î≤îÏúÑ: 0~6ùúã 
* XÏ∂ï Ïù¥Î¶Ñ: time
* YÏ∂ï Ïù¥Î¶Ñ: signal
* Î≤îÎ°Ä: sin‚Å°(ùë•+ùúã) Í∑∏ÎûòÌîÑÎäî ‚Äúsine‚Äù
* Î≤îÎ°Ä: ùëêùëúùë†‚Å°(ùë•) Í∑∏ÎûòÌîÑÎäî ‚Äúcosine‚Äù
* Ïù¥ Îïå Î≤îÎ°ÄÎäî  Í∑∏ÎûòÌîÑÏùò Ïò§Î•∏Ï™Ω ÏÉÅÎã®Ïóê ÏúÑÏπò 
* sinÍ∑∏ÎûòÌîÑÎäî Îπ®Í∞ÑÏÉâ, cosÍ∑∏ÎûòÌîÑÎäî ÌååÎûÄÏÉâÏúºÎ°ú ÌëúÌòÑ


```python
x = np.arange(0, (np.pi)*6, 0.1)
p = np.pi
s = sin(x+p)
c = cos(x)

# sin, cosine Í∑∏Î¶¨Í∏∞
plt.plot(x, s, label='sine', color='red') #Í∑∏Î¶ºÍ≥º Í∞ôÍ≤å ÌïòÎ†§Î©¥ label='sin'
plt.plot(x, c, markersize=10, label='cosine', color='blue') #Í∑∏Î¶ºÍ≥º Í∞ôÍ≤å ÌïòÎ†§Î©¥ label='cos'

# Ï∂ï Î†àÏù¥Î∏î, Ï∂ïÏ†úÎ™©
plt.xlabel('time')
plt.ylabel('signal')
plt.title('Signal Processing')

# Î≤îÎ°Ä
plt.legend(loc='upper right')

plt.show()
```

## 2. sin‚Å°(ùë•+ùúã)  ÏôÄ cos‚Å°(ùë•) Î•º subplotÏúºÎ°ú Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∑∏Î¶¨ÏãúÏò§.

```python
x = np.arange(0, (np.pi)*6, 0.1)
p = np.pi
s = sin(x+p)
c = cos(x)

# sin Í∑∏Î¶¨Í∏∞
plt.subplot(1,2,1)
plt.plot(x, s,label='sine', color='red')
plt.xlabel('time')
plt.ylabel('sin')
plt.xticks(np.arange(0,(np.pi)*6,10))

# cos Í∑∏Î¶¨Í∏∞
plt.subplot(1,2,2)
plt.plot(x, c, label='cosine', color='blue')
plt.xlabel('time')
plt.ylabel('cos')
plt.xticks(np.arange(0,(np.pi)*6,10))

# plot Í∞Ñ Í∞ÑÍ≤© Ï°∞Ï†ï
plt.subplots_adjust(right=1.5)

plt.show()
```

## 3. seoul.csv ÌååÏùºÏóêÎäî ÏÑúÏö∏Ïùò 1908ÎÖÑ 1Ïõî 1ÏùºÎ∂ÄÌÑ∞ 2020ÎÖÑ 12Ïõî 31ÏùºÏùºÍπåÏßÄÏùò ÏùºÎ≥Ñ Í∏∞Ïò® ÏûêÎ£åÍ∞Ä Ï†ÄÏû•ÎêòÏñ¥ÏûàÎã§. Ï†ÑÏ≤¥ Í∏∞Í∞Ñ ÎèôÏïà ÏµúÍ≥†Í∏∞Ïò®Ïóê ÎåÄÌï¥ÏÑú ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¨ÏãúÏò§. (Í≤∞Ï∏°ÏπòÎäî Í≥µÎ∞±ÏúºÎ°ú Ï†ÄÏû•ÎêòÏñ¥ ÏûàÏùå)

```python
with open('seoul.csv', 'r', encoding='cp949') as f:
    seoul = [k for k in csv.reader(f)]
```

```python
seoul = seoul[1:]
```

```python
high = []
for row in seoul :
    if row[-1] !='':
        high.append(float(row[-1]))
        
plt.hist(high, bins=100)
plt.show()
```

## 4. 8ÏõîÏùò ÏµúÍ≥†Í∏∞Ïò®Îßå Ïù¥Ïö©ÌïòÏó¨ ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¨ÏãúÏò§.

```python
aug = []
for row in seoul :
    if row[-1] !='':
        if row[0][5:7] == '08':
            aug.append(float(row[-1]))

plt.hist(aug, bins=20)
plt.show()
```

## 5. 1ÏõîÍ≥º 8ÏõîÏùò ÏµúÍ≥†Í∏∞Ïò®Ïóê ÎåÄÌï¥ÏÑú ÏÉÅÏûêÍ∑∏Î¶ºÏùÑ Í∑∏Î¶¨ÏãúÏò§.

```python
jan = []
for row in seoul :
    if row[-1] !='':
        if row[0][5:7] == '01':
            jan.append(float(row[-1]))
plt.boxplot([jan,aug])
plt.show()
```
