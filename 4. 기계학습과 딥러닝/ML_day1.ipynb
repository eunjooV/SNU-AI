# Data Processing

```python
np.random.seed(156)
```

```python
car = pd.read_csv("./CarPrice.csv")
car.head()
```

## Splitting Training & Test Set

```python
car["price_cut"] = pd.cut(car["price"],
                               bins=[0,10000,20000,30000,40000,np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
car["price_cut"]
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(car, test_size=0.2, random_state=42, stratify=car["price_cut"].values)
```

```python
test_set.head()
```

## Missing Values

```python
car.iloc[0]['wheelbase']
```

```python
car['wheelbase'].replace([88.6], np.nan, inplace=True)
```

```python
car['wheelbase'].fillna(car['wheelbase'].median(), inplace=True)
```

```python
car
```

## Categorical Attribute

```python
car_cat = car[['drivewheel']]
```

```python
car_cat
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(car_cat)
housing_cat_encoded[:10]
```

```python
from sklearn.preprocessing import OneHotEncoder

hot_encoder = OneHotEncoder()
housing_cat_1hot = hot_encoder.fit_transform(car_cat)
housing_cat_1hot.toarray()
```
